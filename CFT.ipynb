# Выделим некоторые фичи для покупателей из train_purch

```python
# Создадим признак - "среднее количество товаров в чеке для каждого покупателя"

AVG_purch_pcs = train_purch.groupby(['client_id', 'transaction_id'], as_index = False).size()\
    .groupby('client_id', as_index = False)['size'].mean()\
    .rename(columns = {'size':'AVG_purch_pcs'})
```

```python
AVG_purch_pcs
```

```python
# Добавим в таблицу с клиентами столбец со значением среднего кол-ва единиц товаров в чеке

df_processed = train.merge(AVG_purch_pcs, on='client_id')
```

```python
df_processed
```

```python
# Создание признака среднего чека в деньгах (сколько в среднем денег тратил клиент за одно посещение магазина)

AVG_purch_currency = (train_purch.groupby(['client_id', 'transaction_id'], as_index=False)['purchase_sum'].mean())\
    .groupby(['client_id'], as_index=False)['purchase_sum'].mean()\
    .rename(columns = {'purchase_sum': 'AVG_purch_currency'})
```

```python
AVG_purch_currency
```

```python
AVG_purch_currency[AVG_purch_currency['client_id'] == 'e99e6fabb9']
```

```python
# Добавление признака AVG_purch_currency в таблицу df_processed

df_processed = df_processed.merge(AVG_purch_currency, on='client_id')
```

```python
df_processed
```

```python
# Подсчет общего количества чеков(посещений) для каждого покупателя

visit_count = train_purch.groupby(['client_id', 'transaction_id'], as_index=False).size()\
    .groupby(['client_id'], as_index=False).size()\
    .rename(columns={'size':'visit_count'})
visit_count

```

```python
# Добавление столбца Visit_count в матрицу признаков 

df_processed = df_processed.merge(visit_count, on='client_id')
df_processed
```

```python
# Посчитаем средний интервал времени между покупками для каждого покупателя

# Сначала создадим таблицу содержащую информацию о датах посещений магазина для каждого покупателя, 
# и преобразуем столбец transaction_datetime к формату datetime. Удалим столбец 'transaction_datetime'


date = train_purch[['client_id', 'transaction_id', 'transaction_datetime']]
date['transaction_datetime'] = pd.to_datetime(date['transaction_datetime'])
date['trans_date'] = date['transaction_datetime'].dt.date
date = date.drop(columns=['transaction_datetime'], axis=1)
date.head(15)
```

```python
# добавим в таблицу date столбец с количеством посещений из таблицы признаков. Это нужно для расчета среднего 
# интервала между посещениями

date = date.merge(df_processed[['client_id', 'visit_count']], on='client_id')
date
```

```python
# создадим еще два столбца: first_visit - день первого посещения, last_visit - день последнего посещения
# для удобства сначала создадим отдельные датафреймы, а потом смерджим их к date
first_visit = date.groupby('client_id', as_index=False)['trans_date'].min().rename(columns={'trans_date':'first_visit'})
last_visit = date.groupby('client_id', as_index=False)['trans_date'].max().rename(columns={'trans_date':'last_visit'})

date = date.merge(first_visit, on='client_id') \
    .merge(last_visit, on='client_id')
```

```python
# удалим колонку trans_date и все дубликаты
date = date.drop_duplicates(subset = ['client_id'])
date = date.drop(columns=['trans_date'], axis=1)
date = date.reset_index()
date
```

```python
# Теперь добавим столбец "частота посещения магазина", посчитаем этот параметр для каждого посетителя
# как количество дней между первым и последним днем посещения, деленное на количество посещений

date['visit_frequency'] = ((date.last_visit - date.first_visit)/date.visit_count).dt.days
date
```

```python
# Наконец добавим признак visit_frequency к основной матрице признаков

df_processed = df_processed.merge(date[['client_id', 'visit_frequency']], on='client_id')
df_processed
```

```python
# Теперь добавим признаки из файла clients2

df_processed = df_processed.merge(clients2[['client_id', 'age', 'gender']], on='client_id')
df_processed
```
