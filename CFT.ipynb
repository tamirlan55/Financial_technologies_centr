# Ð’Ñ‹Ð´ÐµÐ»Ð¸Ð¼ Ð½ÐµÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ðµ Ñ„Ð¸Ñ‡Ð¸ Ð´Ð»Ñ Ð¿Ð¾ÐºÑƒÐ¿Ð°Ñ‚ÐµÐ»ÐµÐ¹ Ð¸Ð· train_purch

```python
# Ð¡Ð¾Ð·Ð´Ð°Ð´Ð¸Ð¼ Ð¿Ñ€Ð¸Ð·Ð½Ð°Ðº - "ÑÑ€ÐµÐ´Ð½ÐµÐµ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ñ‚Ð¾Ð²Ð°Ñ€Ð¾Ð² Ð² Ñ‡ÐµÐºÐµ Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ Ð¿Ð¾ÐºÑƒÐ¿Ð°Ñ‚ÐµÐ»Ñ"

AVG_purch_pcs = train_purch.groupby(['client_id', 'transaction_id'], as_index = False).size()\
    .groupby('client_id', as_index = False)['size'].mean()\
    .rename(columns = {'size':'AVG_purch_pcs'})
```

```python
AVG_purch_pcs
```

```python
# Ð”Ð¾Ð±Ð°Ð²Ð¸Ð¼ Ð² Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñƒ Ñ ÐºÐ»Ð¸ÐµÐ½Ñ‚Ð°Ð¼Ð¸ ÑÑ‚Ð¾Ð»Ð±ÐµÑ† ÑÐ¾ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸ÐµÐ¼ ÑÑ€ÐµÐ´Ð½ÐµÐ³Ð¾ ÐºÐ¾Ð»-Ð²Ð° ÐµÐ´Ð¸Ð½Ð¸Ñ† Ñ‚Ð¾Ð²Ð°Ñ€Ð¾Ð² Ð² Ñ‡ÐµÐºÐµ

df_processed = train.merge(AVG_purch_pcs, on='client_id')
```

```python
df_processed
```

```python
# Ð¡Ð¾Ð·Ð´Ð°Ð½Ð¸Ðµ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ° ÑÑ€ÐµÐ´Ð½ÐµÐ³Ð¾ Ñ‡ÐµÐºÐ° Ð² Ð´ÐµÐ½ÑŒÐ³Ð°Ñ… (ÑÐºÐ¾Ð»ÑŒÐºÐ¾ Ð² ÑÑ€ÐµÐ´Ð½ÐµÐ¼ Ð´ÐµÐ½ÐµÐ³ Ñ‚Ñ€Ð°Ñ‚Ð¸Ð» ÐºÐ»Ð¸ÐµÐ½Ñ‚ Ð·Ð° Ð¾Ð´Ð½Ð¾ Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ðµ Ð¼Ð°Ð³Ð°Ð·Ð¸Ð½Ð°)

AVG_purch_currency = (train_purch.groupby(['client_id', 'transaction_id'], as_index=False)['purchase_sum'].mean())\
    .groupby(['client_id'], as_index=False)['purchase_sum'].mean()\
    .rename(columns = {'purchase_sum': 'AVG_purch_currency'})
```

```python
AVG_purch_currency
```

```python
AVG_purch_currency[AVG_purch_currency['client_id'] == 'e99e6fabb9']
```

```python
# Ð”Ð¾Ð±Ð°Ð²Ð»ÐµÐ½Ð¸Ðµ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ° AVG_purch_currency Ð² Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñƒ df_processed

df_processed = df_processed.merge(AVG_purch_currency, on='client_id')
```

```python
df_processed
```

```python
# ÐŸÐ¾Ð´ÑÑ‡ÐµÑ‚ Ð¾Ð±Ñ‰ÐµÐ³Ð¾ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð° Ñ‡ÐµÐºÐ¾Ð²(Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ð¹) Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ Ð¿Ð¾ÐºÑƒÐ¿Ð°Ñ‚ÐµÐ»Ñ

visit_count = train_purch.groupby(['client_id', 'transaction_id'], as_index=False).size()\
    .groupby(['client_id'], as_index=False).size()\
    .rename(columns={'size':'visit_count'})
visit_count

```

```python
# Ð”Ð¾Ð±Ð°Ð²Ð»ÐµÐ½Ð¸Ðµ ÑÑ‚Ð¾Ð»Ð±Ñ†Ð° Visit_count Ð² Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ñƒ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð² 

df_processed = df_processed.merge(visit_count, on='client_id')
df_processed
```

```python
# ÐŸÐ¾ÑÑ‡Ð¸Ñ‚Ð°ÐµÐ¼ ÑÑ€ÐµÐ´Ð½Ð¸Ð¹ Ð¸Ð½Ñ‚ÐµÑ€Ð²Ð°Ð» Ð²Ñ€ÐµÐ¼ÐµÐ½Ð¸ Ð¼ÐµÐ¶Ð´Ñƒ Ð¿Ð¾ÐºÑƒÐ¿ÐºÐ°Ð¼Ð¸ Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ Ð¿Ð¾ÐºÑƒÐ¿Ð°Ñ‚ÐµÐ»Ñ

# Ð¡Ð½Ð°Ñ‡Ð°Ð»Ð° ÑÐ¾Ð·Ð´Ð°Ð´Ð¸Ð¼ Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñƒ ÑÐ¾Ð´ÐµÑ€Ð¶Ð°Ñ‰ÑƒÑŽ Ð¸Ð½Ñ„Ð¾Ñ€Ð¼Ð°Ñ†Ð¸ÑŽ Ð¾ Ð´Ð°Ñ‚Ð°Ñ… Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ð¹ Ð¼Ð°Ð³Ð°Ð·Ð¸Ð½Ð° Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ Ð¿Ð¾ÐºÑƒÐ¿Ð°Ñ‚ÐµÐ»Ñ, 
# Ð¸ Ð¿Ñ€ÐµÐ¾Ð±Ñ€Ð°Ð·ÑƒÐµÐ¼ ÑÑ‚Ð¾Ð»Ð±ÐµÑ† transaction_datetime Ðº Ñ„Ð¾Ñ€Ð¼Ð°Ñ‚Ñƒ datetime. Ð£Ð´Ð°Ð»Ð¸Ð¼ ÑÑ‚Ð¾Ð»Ð±ÐµÑ† 'transaction_datetime'


date = train_purch[['client_id', 'transaction_id', 'transaction_datetime']]
date['transaction_datetime'] = pd.to_datetime(date['transaction_datetime'])
date['trans_date'] = date['transaction_datetime'].dt.date
date = date.drop(columns=['transaction_datetime'], axis=1)
date.head(15)
```

```python
# Ð´Ð¾Ð±Ð°Ð²Ð¸Ð¼ Ð² Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñƒ date ÑÑ‚Ð¾Ð»Ð±ÐµÑ† Ñ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾Ð¼ Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ð¹ Ð¸Ð· Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñ‹ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð². Ð­Ñ‚Ð¾ Ð½ÑƒÐ¶Ð½Ð¾ Ð´Ð»Ñ Ñ€Ð°ÑÑ‡ÐµÑ‚Ð° ÑÑ€ÐµÐ´Ð½ÐµÐ³Ð¾ 
# Ð¸Ð½Ñ‚ÐµÑ€Ð²Ð°Ð»Ð° Ð¼ÐµÐ¶Ð´Ñƒ Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸ÑÐ¼Ð¸

date = date.merge(df_processed[['client_id', 'visit_count']], on='client_id')
date
```

```python
# ÑÐ¾Ð·Ð´Ð°Ð´Ð¸Ð¼ ÐµÑ‰Ðµ Ð´Ð²Ð° ÑÑ‚Ð¾Ð»Ð±Ñ†Ð°: first_visit - Ð´ÐµÐ½ÑŒ Ð¿ÐµÑ€Ð²Ð¾Ð³Ð¾ Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ñ, last_visit - Ð´ÐµÐ½ÑŒ Ð¿Ð¾ÑÐ»ÐµÐ´Ð½ÐµÐ³Ð¾ Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ñ
# Ð´Ð»Ñ ÑƒÐ´Ð¾Ð±ÑÑ‚Ð²Ð° ÑÐ½Ð°Ñ‡Ð°Ð»Ð° ÑÐ¾Ð·Ð´Ð°Ð´Ð¸Ð¼ Ð¾Ñ‚Ð´ÐµÐ»ÑŒÐ½Ñ‹Ðµ Ð´Ð°Ñ‚Ð°Ñ„Ñ€ÐµÐ¹Ð¼Ñ‹, Ð° Ð¿Ð¾Ñ‚Ð¾Ð¼ ÑÐ¼ÐµÑ€Ð´Ð¶Ð¸Ð¼ Ð¸Ñ… Ðº date
first_visit = date.groupby('client_id', as_index=False)['trans_date'].min().rename(columns={'trans_date':'first_visit'})
last_visit = date.groupby('client_id', as_index=False)['trans_date'].max().rename(columns={'trans_date':'last_visit'})

date = date.merge(first_visit, on='client_id') \
    .merge(last_visit, on='client_id')
```

```python
# ÑƒÐ´Ð°Ð»Ð¸Ð¼ ÐºÐ¾Ð»Ð¾Ð½ÐºÑƒ trans_date Ð¸ Ð²ÑÐµ Ð´ÑƒÐ±Ð»Ð¸ÐºÐ°Ñ‚Ñ‹
date = date.drop_duplicates(subset = ['client_id'])
date = date.drop(columns=['trans_date'], axis=1)
date = date.reset_index()
date
```

```python
# Ð¢ÐµÐ¿ÐµÑ€ÑŒ Ð´Ð¾Ð±Ð°Ð²Ð¸Ð¼ ÑÑ‚Ð¾Ð»Ð±ÐµÑ† "Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð° Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ñ Ð¼Ð°Ð³Ð°Ð·Ð¸Ð½Ð°", Ð¿Ð¾ÑÑ‡Ð¸Ñ‚Ð°ÐµÐ¼ ÑÑ‚Ð¾Ñ‚ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€ Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ Ð¿Ð¾ÑÐµÑ‚Ð¸Ñ‚ÐµÐ»Ñ
# ÐºÐ°Ðº ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð´Ð½ÐµÐ¹ Ð¼ÐµÐ¶Ð´Ñƒ Ð¿ÐµÑ€Ð²Ñ‹Ð¼ Ð¸ Ð¿Ð¾ÑÐ»ÐµÐ´Ð½Ð¸Ð¼ Ð´Ð½ÐµÐ¼ Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ñ, Ð´ÐµÐ»ÐµÐ½Ð½Ð¾Ðµ Ð½Ð° ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð¿Ð¾ÑÐµÑ‰ÐµÐ½Ð¸Ð¹

date['visit_frequency'] = ((date.last_visit - date.first_visit)/date.visit_count).dt.days
date
```

```python
# ÐÐ°ÐºÐ¾Ð½ÐµÑ† Ð´Ð¾Ð±Ð°Ð²Ð¸Ð¼ Ð¿Ñ€Ð¸Ð·Ð½Ð°Ðº visit_frequency Ðº Ð¾ÑÐ½Ð¾Ð²Ð½Ð¾Ð¹ Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ðµ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð²

df_processed = df_processed.merge(date[['client_id', 'visit_frequency']], on='client_id')
df_processed
```
